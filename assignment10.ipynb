```python
import numpy as np

arr = np.array([
    [6, -8, 73, -110],
    [np.nan, -8, 0, 94]
])

print("Original Array")
print(arr)
```

```python
arr = np.nan_to_num(arr, nan=0)

print("\nAfter Replacing NaN with 0")
print(arr)
```

```python
transpose_arr = arr.T

print("\nTranspose Array")
print(transpose_arr)
```

```python
arr3d = np.array([
    [[1, 2], [3, 4]],
    [[5, 6], [7, 8]]
])

print("Original Shape:", arr3d.shape)
```

```python
new_arr = np.moveaxis(arr3d, 0, 2)

print("\nNew Array")
print(new_arr)

print("\nNew Shape:", new_arr.shape)
```

```python
arr = np.array([
    [1, 2, np.nan],
    [4, np.nan, 6],
    [7, 8, 9]
])

print("Original Array")
print(arr)
```

```python
col_mean = np.nanmean(arr, axis=0)

inds = np.where(np.isnan(arr))

arr[inds] = np.take(col_mean, inds[1])

print("\nArray After Replacing NaN")
print(arr)
```

```python
arr = np.array([4, -3, 7, -1, 9, -6])

print("Original Array")
print(arr)

arr[arr < 0] = 0

print("\nUpdated Array")
print(arr)
```
