```python
import numpy as np

# 1. Create NumPy array
arr = np.array([10, 20, 30, 40, 50])
print("Original Array:")
print(arr)

# Convert 1D array to 2D
arr_2d = arr.reshape(1, 5)
print("\n1D to 2D Array:")
print(arr_2d)

# Print Array Attributes
print("\nArray Attributes:")
print("Shape:", arr.shape)
print("Dimension:", arr.ndim)
print("Data Type:", arr.dtype)
print("Item Size:", arr.itemsize)
```

```python
# 2. Create a 3×3 NumPy array of all 9
arr9 = np.full((3, 3), 9)
print("\n3x3 Array of 9:")
print(arr9)
```

```python
# 3. Create a 1D array of 10 evenly spaced values between 25 and 125
even_arr = np.linspace(25, 125, 10)
print("\n10 Evenly Spaced Values:")
print(even_arr)
```

```python
# 4. Convert Python list into NumPy array
lst = [1, 2, 3, 4, 5]
np_arr = np.array(lst)
print("\nList to NumPy Array:")
print(np_arr)
```

```python
# 5. Reverse a 1D NumPy array
arr_rev = np_arr[::-1]
print("\nReversed Array:")
print(arr_rev)
```

```python
# 6. Create a 4×4×3 array and extract value at second set, first row and last column
arr3d = np.arange(48).reshape(4, 4, 3)
print("\n4x4x3 Array:")
print(arr3d)

value = arr3d[1, 0, 2]
print("\nSecond Set, First Row, Last Column:")
print(value)
```

```python
# 7. Create a 4×4 array and extract odd rows and even columns
arr4 = np.arange(1, 17).reshape(4, 4)
print("\n4x4 Array:")
print(arr4)

result = arr4[::2, 1::2]
print("\nOdd Rows and Even Columns:")
print(result)
```

```python

# 8. Slice first two rows and first two columns of second set from 4x4x3 array
slice_arr = arr3d[1, :2, :2]
print("\nFirst Two Rows and First Two Columns of Second Set:")
print(slice_arr)
```

```python
# 9. Replace all odd numbers with -1 using for loop
arr_replace = np.array([[23, 56, 78, 93],
                        [71, 82, 13, 24]])

for i in range(arr_replace.shape[0]):
    for j in range(arr_replace.shape[1]):
        if arr_replace[i, j] % 2 != 0:
            arr_replace[i, j] = -1

print("\nReplace Odd Numbers with -1:")
print(arr_replace)
```

```python
# 10. Get indices of non-zero elements
arr_nonzero = np.array([1, 0, 2, 0, 3, 0, 4])

indices = np.nonzero(arr_nonzero)
print("\nIndices of Non-Zero Elements:")
print(indices)
```

```python

# 11. Arithmetic Operations
a = np.array([10, 20, 30])
b = np.array([5, 10, 15])

print("\nAddition:")
print(a + b)

print("\nMultiplication:")
print(a * b)
```

```python
# 12. Dot Product
arr1 = np.array([15, 20, 25])
arr2 = np.array([10, 40, 37])

dot_product = np.dot(arr1, arr2)

print("\nDot Product:")
print(dot_product)
```
