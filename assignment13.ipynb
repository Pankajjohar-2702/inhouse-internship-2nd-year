```python
import numpy as np

arr = np.array([[6, -8, 73, -110],
                [np.nan, -8, 0, 94]])

print("Original Array:")
print(arr)

# Replace NaN with 0
arr = np.nan_to_num(arr, nan=0)

print("\nAfter Replacing NaN with 0:")
print(arr)

# Interchange rows and columns (Transpose)
arr_transpose = arr.T

print("\nAfter Interchanging Rows and Columns:")
print(arr_transpose)
```

```python
import numpy as np

arr = np.array([[10, np.nan, 30],
                [20, 25, np.nan],
                [30, 35, 50]])

print("Original Array:")
print(arr)

# Column means ignoring NaN
col_mean = np.nanmean(arr, axis=0)

# Replace NaN with column mean
inds = np.where(np.isnan(arr))
arr[inds] = np.take(col_mean, inds[1])

print("\nAfter Replacing NaN with Column Average:")
print(arr)
```

```python
import numpy as np

arr = np.array([[6, -8, 73, -110],
                [12, -5, 0, 94]])

print("Original Array:")
print(arr)

arr[arr < 0] = 0

print("\nAfter Replacing Negative Values with 0:")
print(arr)
```

```python
import numpy as np
from scipy import stats

arr1 = np.array([[3, 4],
                 [5, 6]])

arr2 = np.array([[1, 0],
                 [7, 8]])

# Average of arrays
avg = (arr1 + arr2) / 2

print("Average Array:")
print(avg)

# Combine arrays
combined = np.concatenate((arr1.flatten(),
                           arr2.flatten()))

print("\nMean:", np.mean(combined))
print("Median:", np.median(combined))

mode_result = stats.mode(combined, keepdims=True)

print("Mode:", mode_result.mode[0])
```

```python
# x−2y+3z=9
# −x+3y−z=−6
# 2x−5y+5z=17
import numpy as np

A = np.array([[1, -2, 3],
              [-1, 3, -1],
              [2, -5, 5]])

B = np.array([9, -6, 17])

solution = np.linalg.solve(A, B)

print("x =", solution[0])
print("y =", solution[1])
print("z =", solution[2])
```

```python
import matplotlib.pyplot as plt

subjects = ["Math", "Physics", "Chemistry", "English", "CS"]

semester1 = [75, 68, 80, 72, 85]
semester2 = [82, 75, 84, 78, 90]

plt.figure(figsize=(8,5))

plt.plot(subjects,
         semester1,
         marker='o',
         linestyle='--',
         label='Semester 1')

plt.plot(subjects,
         semester2,
         marker='s',
         linestyle='-',
         label='Semester 2')

plt.title("Semester Result Comparison")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.grid(True)
plt.legend()

plt.show()
```
