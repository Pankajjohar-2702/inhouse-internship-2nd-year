```python
# Tuple Example

numbers = (10, 20, 30, 40)

print("Tuple:", numbers)

print("First Element:", numbers[0])

print("Length of Tuple:", len(numbers))
```

```python
# Dictionary Example

student = {
    "name": "Pankaj",
    "class": "B.Tech",
    "marks": 85
}

print("Student Details:")
print(student)

print("Name:", student["name"])

# Adding new item
student["age"] = 20

print("Updated Dictionary:")
print(student)
```

```python
# Set Example

fruits = {"apple", "banana", "mango"}

print("Set:", fruits)

# Adding element
fruits.add("orange")

print("Updated Set:", fruits)
```

```python
# Function for Basic Math Operations

def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

def multiply(a, b):
    return a * b

def divide(a, b):
    if b != 0:
        return a / b
    else:
        return "Division by zero not possible"

# Taking input from user
a = float(input("Enter first number: "))
b = float(input("Enter second number: "))

print("Addition:", add(a, b))
print("Subtraction:", subtract(a, b))
print("Multiplication:", multiply(a, b))
print("Division:", divide(a, b))
```

```python
# Program to Check Palindrome Number

num = int(input("Enter a number: "))

original_num = num
reverse_num = 0

while num > 0:
    digit = num % 10
    reverse_num = reverse_num * 10 + digit
    num = num // 10

print("Reverse Number:", reverse_num)

if original_num == reverse_num:
    print(original_num, "is a Palindrome Number")
else:
    print(original_num, "is not a Palindrome Number")
```
