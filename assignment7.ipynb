```python
import pandas as pd

data = {
    'Name': ['Pankaj', 'Rahul', 'Amit'],
    'Marks': [90, 85, 88]
}

df = pd.DataFrame(data)

# Using iterrows()
for index, row in df.iterrows():
    print(row['Name'], row['Marks'])
```

```python
import pandas as pd

data = {
    'Name': ['Pankaj', 'Rahul'],
    'Marks': [90, 85]
}

df = pd.DataFrame(data)

new_row = pd.DataFrame({
    'Name': ['Amit'],
    'Marks': [88]
})

df1 = pd.concat([df.iloc[:1], new_row, df.iloc[1:]]).reset_index(drop=True)

print(df1)
```
