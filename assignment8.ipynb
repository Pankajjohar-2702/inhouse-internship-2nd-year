```python
import pandas as pd

df1 = pd.DataFrame({
    'ID': [1, 2, 3, 4],
    'Name': ['Aman', 'Ravi', 'Simran', 'Karan']
})

df2 = pd.DataFrame({
    'ID': [2, 3, 4, 5],
    'Marks': [85, 90, 88, 76]
})

print("DF1")
print(df1)

print("\nDF2")
print(df2)
```

```python
inner_merge = pd.merge(df1, df2, on='ID', how='inner')

print(inner_merge)
```

```python
left_join = pd.merge(df1, df2, on='ID', how='left')

print(left_join)
```

```python
right_join = pd.merge(df1, df2, on='ID', how='right')

print(right_join)
```

```python
df1_index = df1.set_index('ID')
df2_index = df2.set_index('ID')

join_result = df1_index.join(df2_index)

print(join_result)
```

```python
df3 = pd.DataFrame({
    'ID': [1, 2, 1, 2],
    'Subject': ['Math', 'Math', 'Science', 'Science'],
    'Marks': [80, 85, 78, 88]
})

df4 = pd.DataFrame({
    'ID': [1, 2, 1, 2],
    'Subject': ['Math', 'Math', 'Science', 'Science'],
    'Teacher': ['A', 'B', 'C', 'D']
})

multi_merge = pd.merge(df3, df4, on=['ID', 'Subject'])

print(multi_merge)
```

```python
import pandas as pd

df_a = pd.DataFrame({
    'ID': [1, 2],
    'Name': ['Aman', 'Ravi']
})

df_b = pd.DataFrame({
    'ID': [3, 4],
    'Name': ['Simran', 'Karan']
})

df_c = pd.DataFrame({
    'ID': [1, 2, 3, 4],
    'City': ['Delhi', 'Mumbai', 'Pune', 'Jaipur']
})
concat_df = pd.concat([df_a, df_b], ignore_index=True)

print(concat_df)
```

```python
final_df = pd.merge(concat_df, df_c, on='ID')

print(final_df)

# Comparison of join() and merge() in pandas


# ---------join()----------                   --------- merge()----------
# Used mainly for joining on index	    |       Used for joining on columns
# Simpler syntax	                    |       More powerful and flexible
# Default join is left join             |	    Default join is inner join
# Best for index-based operations	    |       Best for SQL-like operations
# Uses DataFrame.join()	                |       Uses pd.merge()
```
